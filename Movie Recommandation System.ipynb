**Now we are going to first convert the string into the list**

```python
import ast
ast.literal_eval('[{"id": 28, "name": "Action"}, {"id": 12, "name": "Adventure"}, {"id": 14, "name": "Fantasy"}, {"id": 878, "name": "Science Fiction"}]')
```

```python
def convert(obj):
    L = []
    for i in ast.literal_eval(obj):
        L.append(i['name'])
    return L
```

```python
def convert1(obj):
    L = []
    counter = 0
    for i in ast.literal_eval(obj):
        if counter != 3:
            L.append(i['name'])
            counter += 1
        else:
            break
    return L
```

```python
def director(obj):
    L = []
    for i in ast.literal_eval(obj):
        if i['job'] == 'Director':
            L.append(i['name'])
            break
    return L
```

**Here we gone use the nltk library for the steming.  
steming means it convert ['loved','love','loving']  to   ['love','love','love']  
it will help use when we convert the tags into the matrix or numpy array.**

```python
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
ps.stem('loving')
```

```python
ps.stem('loved')
```

```python
def stem(text):
    y = []
    for i in text.split():
        y.append(ps.stem(i))
        
    return " ".join(y)
```

**Here we gone apply first convert the hole string into list then split it and apply the stem on each words and then again convert it into the string and then return it.**

# Why we use cosine distance instead of euclidean distance?

**Becouse in the higher dimenssions the euclidean distance is not work properly.
so thats why we use cosine distance.**

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
cosine_similarity(vector)
```

```python
cosine_similarity(vector).shape
```

**Here we done the distance calculation for 1 movie to other 4806 movies.  
from this we use cosine_similarity class this is used for the distance calculation it can vary from 0 to 1.  
Higher the value of cosine_similarity value higher the similarity.**

```python
similarity = cosine_similarity(vector)
```

**Here we sort the distances.**

**Here there is one problem when we apply sort function we show it in descending order but our index position is lose.

```python
similarity[0]
```

**So for getting over to this type of problem we use enumerate function.**

```python
list(enumerate(similarity[0]))
```

**Now we can use sort**

```python
sorted(list(enumerate(similarity[0])), reverse = True, key = lambda x:x[1])
```

**Here we use "key = lambda x:x[1]" this function becouse if we do not use it will sort according to the 1,2,3,4 i.e the id of the movies.**

```python
sorted(list(enumerate(similarity[0])), reverse = True, key = lambda x:x[1])[1:6]
```

```python
def recommend(movie):
    movies_index = new[new['title'] == movie].index[0]
    distances = similarity[movies_index]
    movies_list = sorted(list(enumerate(distances)), reverse = True, key = lambda x:x[1])[1:11]
    
    for i in movies_list:
        print(new.iloc[i[0]].title)
```

**Here we made the recommandation def functio for the recommendation system this is the code for the recommendation.**

```python
recommend('Avatar')
```

```python
recommend('Batman Begins')
```
